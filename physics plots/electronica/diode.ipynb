```python
#imports
import numpy as np 
import matplotlib.pyplot as plt 
import prettytable as tbl
```

```python
# dioda de siliciu inversata

#volti
Uis = np.array([-0.3,-0.6, -0.9, -1.2, -1.5, -1.8, -2, -8.8, -8.9, -8.91, -8.92, -8.93, -8.94, -8.95])
# miliamp * 10
Iis = np.array([0   , 0  , 0   , 0   , 0   , 0   , 0 , -0.2, -0.8, -2.1 , -3.6 , -5.4 , -7.3 , -9.3])

Iis /= 10.0

tabel = tbl.PrettyTable()

tabel.add_column("Si - I (mA)", np.round(Iis, 3))
tabel.add_column("Si - U (V)", np.round(Uis, 3))
print(tabel)
```

```python
# dioda de siliciu

#volti
Uds = np.array([0.19, 0.39, 0.53, 0.57, 0.59, 0.60, 0.61, 0.62, 0.63, 0.63, 0.64, 0.64])
# miliamp * 10
Ids = np.array([0.1 , 0.1 , 0.1 , 0.15, 0.2 , 0.6 , 1.3 , 2.1 , 3.2 , 4.4 , 5.5 , 6.8])

Ids /= 10.0

tabel = tbl.PrettyTable()

tabel.add_column("Si - I (mA)", np.round(Ids, 3))
tabel.add_column("Si - U (V)", np.round(Uds, 3))
print(tabel)
```

```python
# dioda de germaniu

#miliamp * 10
Idg = np.array([0.1 ,0.1  , 0.7 ,1.6 ,2.8   ,4.2   ,5.6   ,   7.2,     9])
#volti
Udg = np.array([0.05, 0.07, 0.09, 0.1, 0.115, 0.123, 0.131, 0.138, 0.144])

Idg /= 10

tabel = tbl.PrettyTable()
tabel.add_column("Ge - I (mA)", np.round(Idg, 3))
tabel.add_column("Ge - U (V)", np.round(Udg, 3))
print(tabel)
```

```python
# LED

#miliamp * 10
Idl = np.array([0,0,0,0,0,0,0,0.2,0.25,0.4,0.7,1,2])
#volti
Udl = np.array([0.19,0.39,0.6,0.8,1,1.2,1.4,1.5,1.6,1.65,1.67,1.68,1.7])

Idl /= 10

tabel = tbl.PrettyTable()
tabel.add_column("LED - I (mA)", np.round(Idl, 3))
tabel.add_column("LED - U (V)", np.round(Udl, 3))
print(tabel)

```

```python
fig = plt.figure()
ax = plt.axes()
ax.plot(Uds, Ids, linestyle = '-', marker = 'o', markersize = 3, color = 'r', label = 'Si')
ax.plot(Uis, Iis, linestyle = '-', marker = 'o', markersize = 3, color = 'r', label = 'Si inversata')
ax.plot(Udg, Idg, linestyle = '-', marker = 'o', markersize = 3, color = 'b', label = 'Ge')
ax.plot(Udl, Idl, linestyle = '-', marker = 'o', markersize = 3, color = 'g', label = 'LED')
ax.axhline(y = 0, color = 'k', lw = 3)
ax.axvline(x = 0, color = 'k', lw = 3)
ax.grid(which = 'major', axis = 'both', color='#dddddd', linestyle='-')
ax.grid(which = 'minor', axis = 'both', color='#eeeeee', linestyle=':')
ax.minorticks_on()
ax.legend(loc='best')
plt.xlabel('U (V)', fontsize = 18)
plt.ylabel('I (mA)', fontsize = 18)
ax.xaxis.labelpad = 20
ax.yaxis.labelped = 20
fig.tight_layout()
plt.show()
```
