```python
#imports
import numpy as np 
import matplotlib.pyplot as plt 
import prettytable as tbl
```

```python
#Sursa ideala de tensiune

Rsitt = np.array([2000,50, 27, 5.1])
Usit = np.array([2.7,4.96, 4.96, 4.9])
#miliamp
Isit = np.array([1.35,99.2, 183.7, 960.8])

Rsite = Usit / Isit * 1000

tabel = tbl.PrettyTable()

tabel.add_column("I (mA)", np.round(Isit, 3))
tabel.add_column("U (V)", np.round(Usit, 3))
tabel.add_column("R (Ω) teoretic", np.round(Rsitt, 3))
tabel.add_column("R (Ω) experimental", np.round(Rsite, 3))

print('Ugol = 4.97V')
print(tabel)

fig = plt.figure()
ax = plt.axes()
ax.plot(Isit, Usit, linestyle = '-', marker = 'o', markersize = 3, color = 'r', label = 'Sursa ideala de tensiune')
ax.axhline(y = 0, color = 'k', lw = 3)
ax.axvline(x = 0, color = 'k', lw = 3)
ax.grid(which = 'major', axis = 'both', color='#dddddd', linestyle='-')
ax.grid(which = 'minor', axis = 'both', color='#eeeeee', linestyle=':')
ax.minorticks_on()
ax.legend(loc='best')
plt.xlabel('I (mA)', fontsize = 18)
plt.ylabel('U (V)', fontsize = 18)
ax.xaxis.labelpad = 20
ax.yaxis.labelped = 20
fig.tight_layout()
plt.show()
```

```python
#Sursa ideala de curent

Rsict = np.array([10000, 4700,3000, 1500, 750])
Usic = np.array([15.13, 13.67, 12.3, 7.85, 3.97])
#miliamp
Isic = np.array([1.513, 2.91, 4.1, 5.23, 5.29])

Rsice = Usic / Isic * 1000

tabel = tbl.PrettyTable()

tabel.add_column("I (mA)", np.round(Isic, 3))
tabel.add_column("U (V)", np.round(Usic, 3))
tabel.add_column("R (Ω) teoretic", np.round(Rsict, 3))
tabel.add_column("R (Ω) experimental", np.round(Rsice, 3))

print(tabel)

fig = plt.figure()
ax = plt.axes()
ax.plot(Isic, Usic, linestyle = '-', marker = 'o', markersize = 3, color = 'g', label = 'Sursa ideala de curent')
ax.axhline(y = 0, color = 'k', lw = 3)
ax.axvline(x = 0, color = 'k', lw = 3)
ax.grid(which = 'major', axis = 'both', color='#dddddd', linestyle='-')
ax.grid(which = 'minor', axis = 'both', color='#eeeeee', linestyle=':')
ax.minorticks_on()
ax.legend(loc='best')
plt.xlabel('I (mA)', fontsize = 18)
plt.ylabel('U (V)', fontsize = 18)
ax.xaxis.labelpad = 20
ax.yaxis.labelped = 20
fig.tight_layout()
plt.show()
```
